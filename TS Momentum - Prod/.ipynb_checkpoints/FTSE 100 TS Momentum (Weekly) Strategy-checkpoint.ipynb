```python
import pandas as pd
import numpy as np
import yfinance as yf
from hurst import compute_Hc
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

import warnings
warnings.filterwarnings('ignore')
```

```python
ftse_100_data = yf.download('^FTSE', start='2021-01-01')
```

```python
ftse_100_data.drop(['Open', 'High', 'Low', 'Close', 'Volume'], axis=1,inplace=True)
ftse_100_data.index = pd.to_datetime(ftse_100_data.index)
```

```python
ftse_100_weekly_close = ftse_100_data.asfreq('w', method='ffill')
```

```python
ftse_100_data['Adj Close']['2022-01':]
```

```python
ftse_100_weekly_close.loc['2023-06-11'] = 7562.36
```

```python
ftse_100_weekly_close
```

```python
ftse_100_weekly_close.index = pd.to_datetime(ftse_100_weekly_close.index)
```

```python
ftse_100_weekly_close
```

```python
ftse_100_weekly_close['Previous 2 weeks returns'] = ftse_100_weekly_close.pct_change(2)
```

```python
ftse_100_weekly_close['Next Week Returns'] = ftse_100_weekly_close['Adj Close'].pct_change(1).shift(-1)
```

```python
ftse_100_weekly_close = ftse_100_weekly_close.iloc[2:]
```

```python
ftse_100_weekly_close['positions'] = np.where(ftse_100_weekly_close['Previous 2 weeks returns'] > 0, 1, -1)
```

```python
ftse_100_weekly_close['strategy_returns'] = ftse_100_weekly_close['positions'] * ftse_100_weekly_close['Next Week Returns']
```

```python
ftse_100_weekly_close['cumulative strategy returns'] = (ftse_100_weekly_close['strategy_returns'] + 1).cumprod()
```

```python
ftse_100_weekly_close['cumulative strategy returns'].plot(figsize=(10, 7))
# Set title and labels for the plot
plt.title('Strategy Returns', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Cumulative Returns', fontsize=12)
plt.show()
```

```python
#################################### Signals ######################################################
```

```python
def plot_signal(portfolio):
    ftse_100_weekly_close['Adj Close'].plot(color='blue')    
    plt.fill_between(ftse_100_weekly_close.index, ftse_100_weekly_close['Adj Close'], where=(ftse_100_weekly_close.positions==1), facecolor='g', alpha=0.2)
    plt.fill_between(ftse_100_weekly_close.index, ftse_100_weekly_close['Adj Close'], where=(ftse_100_weekly_close.positions==-1), facecolor='r', alpha=0.2)
    plt.legend()    
    plt.ylabel('Adj Close',color='r')
    plt.show()
plot_signal(ftse_100_weekly_close)
```

```python
#################################### Max Drawdown ##############################################################
```

```python
running_max = np.maximum.accumulate(ftse_100_weekly_close['cumulative strategy returns'].dropna())
running_max[running_max < 1] = 1
# Calculate the percentage drawdown
drawdown = (ftse_100_weekly_close['cumulative strategy returns'])/running_max - 1
max_dd = drawdown.min()*100
print('The maximum drawdown is %.2f' % max_dd)
# Plot the drawdowns
drawdown.plot(color='r',figsize=(10,10))
plt.ylabel('Drawdown')
plt.fill_between(drawdown.index, drawdown, color='red')
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.2)
plt.show()
```

```python
#################################### Buy and Hold FTSE 100 ######################################################
```

```python
Buy_Hold_Returns = ftse_100_weekly_close['Adj Close']
Buy_Hold_Returns = Buy_Hold_Returns.pct_change()
Buy_Hold_Cum_Returns = (Buy_Hold_Returns+1).cumprod()
```

```python
Buy_Hold_Cum_Returns.plot(figsize=(10, 7))

# Set title and labels for the plot
plt.title('Buy and Hold Strategy Returns', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Cumulative Returns', fontsize=12)
plt.show()
```

```python
###################### Long returns #################################
ftse_100_weekly_close['long_cum_returns'] = np.where(ftse_100_weekly_close['positions'] == 1, ftse_100_weekly_close['positions'] * ftse_100_weekly_close['Next Week Returns'], 0)
ftse_100_weekly_close['long_cum_returns'] = (ftse_100_weekly_close['long_cum_returns'] + 1).cumprod()
```

```python
###################### Short returns #################################
ftse_100_weekly_close['short_cum_returns'] = np.where(ftse_100_weekly_close['positions'] == -1, ftse_100_weekly_close['positions'] * ftse_100_weekly_close['Next Week Returns'], 0)
ftse_100_weekly_close['short_cum_returns'] = (ftse_100_weekly_close['short_cum_returns'] + 1).cumprod()
```

```python
ftse_100_weekly_close['long_cum_returns'].plot(figsize=(10, 7), color='g')
ftse_100_weekly_close['short_cum_returns'].plot(figsize=(10, 7), color='r')
# Set title and labels for the plot
plt.title('Long/Short Returns', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Cumulative Returns', fontsize=12)
plt.show()
```
